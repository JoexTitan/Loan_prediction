## Imputation Step Using KNNImputer

```python
from sklearn.impute import KNNImputer

impute_knn = KNNImputer(n_neighbors=5)
num = [col for col in loan_df.columns if loan_df[col].dtypes!='O']
new = pd.DataFrame(impute_knn.fit_transform(loan_df[num]))
```

```python
new.columns=['AI','CI','LA','LAT','CH']
new.describe()
```

```python
loan_df['Loan_Amount_Term'] = new['LAT']

loan_df.describe()
```

```python
loan_df['LoanAmount'] = loan_df.LoanAmount.fillna(loan_df['LoanAmount'].mean())
loan_df['Loan_Amount_Term'] = loan_df['Loan_Amount_Term'].fillna(loan_df['Loan_Amount_Term'].median())
```

## Train Test Split

```python
X = loan_df.drop(columns=['Loan_ID','Loan_Status'],axis=1)
Y = loan_df.Loan_Status
```

```python
print(X,Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
# X_train,Y_test has 90% of data and the correlating answers
# X_test,Y_test has 10% of data and no correlated answers
print(X.shape,X_train.shape,X_test.shape)
```

```python
# Let us use the Support Vector Machine Model
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
# Training the SVM Model
classifier.fit(X_train,Y_train)
```

```python
# A) Model Evaluation - accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print(f'Accuracy on the training data: {training_data_accuracy}')
```

```python
# B) Model Evaluation - accuracy score on the test data -> MUCH MORE IMPORTANT
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print(f'Accuracy on the test data: {test_data_accuracy}')
```
